## List Available Datasets / データセット一覧

```python
from rdh.registry import Registry

reg = Registry()
print(f"Registered datasets: {len(reg.all())}\n")
for entry in reg.all():
    print(f"  {entry.name:15s} | {entry.display_name}")
```

## Dataset Details / 詳細表示

```python
entry = reg.get("covla")
print(f"Name       : {entry.display_name}")
print(f"Description: {entry.description}")
print(f"Paper      : {entry.paper_url}")
print(f"Modalities : {', '.join(entry.modalities)}")
print(f"Tasks      : {', '.join(entry.tasks)}")
print(f"License    : {entry.license}")
print(f"Size       : {entry.size_hint}")
print(f"HuggingFace: {entry.huggingface_id}")
```

## Search Datasets / 検索

```python
print("=== Navigation datasets ===")
for e in reg.search("navigation"):
    print(f"  {e.name}: {e.display_name}")

print("\n=== Maritime datasets ===")
for e in reg.search("maritime"):
    print(f"  {e.name}: {e.display_name}")

print("\n=== SLAM datasets ===")
for e in reg.search("slam"):
    print(f"  {e.name}: {e.display_name}")
```

## Overview Table / 一覧テーブル

```python
import pandas as pd

rows = []
for e in reg.all():
    rows.append({
        "Name": e.name,
        "Display Name": e.display_name,
        "Modalities": ", ".join(e.modalities),
        "Tasks": ", ".join(e.tasks),
        "License": e.license,
        "Size": e.size_hint,
        "Paper": e.paper_url,
    })

df = pd.DataFrame(rows)
df.style.set_properties(**{"text-align": "left"})
```
